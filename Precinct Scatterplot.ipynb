```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

#import matplotlib.cm as cm


df = pd.read_csv('actes-criminels.csv')


crime_category = df['CATEGORIE']

date = (df['DATE'])

# morning, afternoon, evening
time_of_day = (df['QUART'])

# num of police precincts in area
num_precincts = (df['PDQ'])

long = (df['LONGITUDE'])
lat = (df['LATITUDE'])


value_counts = crime_category.value_counts()

# Print the value counts
print(value_counts)

# Generate a bar plot of a column
crime_category.value_counts().plot(kind='bar')

#___________________________________________________________________________________________________________________


color_column = 'PDQ'

color_dict = {val: plt.cm.get_cmap('viridis')
              (i/len(df[color_column].unique())-1) for i
              , val in enumerate(df[color_column].unique())}

df['PRECINCT_COLOR'] = df[color_column].map(color_dict)


#df['PRECINCT_COLOR'] = df[color_column].apply(lambda x: mcolors.to_hex(norm(x/vmax)))

#___________________________________________________________________________________________________________________


fig, ax = plt.subplots(figsize=(10, 8), dpi=100)


df.plot(kind='scatter', x='LONGITUDE', y='LATITUDE', s=0.002 , c='PRECINCT_COLOR', ax=ax)

#, cmap=cm.get_cmap('viridis')

# Show the plot
plt.show()


```
